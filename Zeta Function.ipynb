## Computing Zeta Riemann Function

### Real Values

### $$\displaystyle \zeta (s)=\sum _{n=1}^{\infty }{\frac {1}{n^{s}}}= 1+\frac 1 {2^s}+\frac 1 {3^s}+\cdots$$

```python
def zeta_sum(n,s):
    S=0
    for i in range(1,n+1):
        S+=1/(i**s)
    return S
```

```python
n=10000
zeta_sum(n,1)
```

```python
l = []
for i in range(1,101):
    l.append(zeta_sum(1000,i))
print(l)
```

### Extension in $\mathbb{C}$

### $$\zeta(s)=\frac{1}{1-2^{1-s}} \sum_{n=1}^{\infty} \frac{(-1)^{n-1}}{n^s}$$

```python
from math import *
import sympy as smp
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')
plt.rcParams.update({
    "figure.facecolor":  '#373e4b',
    "axes.facecolor": '#373e4b',
})
```

```python
n, u, k = smp.symbols('n u k')
s=0.5 + 14j
#s=0.5 + 21.022039638771554993*1j
#s=.5 + 22*1j

def Eta(s):
    S=0
    for n in range(1,10**2):
        S+=(-1)**(n-1)/n**s
    return S

Z=(1/(1-2**(1-s))*Eta(s))

smp.re(Z)+smp.im(Z)*1j

#print((1/(1-2**(1-s))))
```

```python
X = np.linspace(0,100,1000)

S = []
Z = []
for i in range(X.shape[0]):
    S.append(complex(1/2, X[i]))

for i in range(X.shape[0]):
    Z.append(1/(1-2**(1-S[i]))*Eta(S[i]))


print(Z)
```

```python
Z = np.array(Z)

plt.plot(X, Z.real,)
plt.plot(X, Z.imag)
plt.plot(X, np.zeros(X.shape[0]), c='w')

plt.show()
```

# Recherches

$$ \lim_{L \to +\infty} \sum_{n=0}^{L}a_n;\sum_{n=0}^{L}b_n\to \frac{(\sum_{n=0}^{L}a_n)\cdot\sum_{n=0}^{L}b_n}{L+1} $$

```python
import numpy as np
```

```python
L=10000

A=np.random.rand(L)
B=np.random.rand(L)

C=np.array([])
D=np.array([])

s=1


for i in range(1,L+1):
    C=np.append(C,1/i**(2*s))
    D=np.append(D,1/i**s)

print("Opération     :", A.dot(B))
print("Approximation :", np.sum(A)*np.sum(B)/L)

print("Opération     :", C.dot(D))
print("Approximation :", np.sum(C)*np.sum(D)/L)
```
