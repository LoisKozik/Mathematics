```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('pdf', 'png')
plt.rcParams['figure.dpi']= 100
plt.rcParams['savefig.dpi'] = 100
```

### $$y' = y$$
### $$y(0) = 1$$

```python
omega = 1

dt = 0.1
time = np.arange(0, 3.001, dt)

f = lambda t,z: z

z = np.zeros_like(time)

# initial conditions
z[0] = 1
```

# Exact Solution

### $$y(t) = e^t$$

```python
# plot exact solution first
time = np.linspace(0,3)
y_exact = np.exp(time)

plt.plot(time, y_exact, label='Exact')
plt.xlabel('time')
plt.ylabel('displacement')
plt.legend()
plt.grid(True)
plt.show()
```

# Heun's Method

### $$f(t, y) = y$$
### $$t_{i+1}=t_i + h$$

### $$ \widetilde{y}_{i+1} =𝑦_𝑖+ℎ𝑓(𝑡_𝑖, 𝑦_𝑖) $$
### $$ 𝑦_{i+1}=𝑦_𝑖 + \frac{h}{2} \Big (𝑓(𝑡_𝑖, 𝑦_𝑖 )+𝑓(𝑡_{i+1},\widetilde{y}_{i+1}) \Big )  $$

```python
dt = 0.1

time = np.arange(0, 3.001, dt)
zp = np.zeros_like(time)

for idx, t in enumerate(time[:-1]):
    zp[idx+1] = z[idx] + dt*f(t,z[idx])
    z[idx+1] = z[idx] + 0.5*dt*(f(t,z[idx])+f(time[idx+1],zp[idx+1]))

plt.plot(time, z, 'o--', label='Heuns method solution')
plt.xlabel('time')
plt.ylabel('displacement')
plt.legend()
plt.grid(True)
plt.show()
```

```python
time = np.linspace(0, 3)
plt.plot(time, y_exact, label='Exact')

time = np.arange(0, 3.001, dt)
plt.plot(time, z, 'o--', label='Heuns method solution')
plt.xlabel('time')
plt.ylabel('displacement')
plt.legend()
plt.grid(True)
plt.show()
```
