```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.rcParams.update({
    "figure.facecolor":  '#373e4b',
    "axes.facecolor": '#373e4b',
})
```

```python
def f(x):
    return 8*(1/(1+np.exp(3-x)))-3 
```

```python
A = -10
B = 10

epsilon = 0.1 # PrÃ©cison attendue
a = A
b = B
```

```python
while (b-a) > epsilon:
    m = (a+b)/2

    if f(a)*f(m)<=0:
        b = m

    else:
        a = m
```

```python
X = np.linspace(A,B)
plt.plot(X,f(X))
plt.plot(X,np.zeros(X.shape), c="white")
plt.scatter(m,f(m), c='r')
```
