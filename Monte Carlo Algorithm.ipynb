```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.rcParams.update({
    "figure.facecolor":  '#373e4b',
    "axes.facecolor": '#373e4b',
})
```

```python
N = 10000

X = np.random.rand(N)
Y = np.random.rand(N)

X = 2*X-1
Y = 2*Y-1
```

```python
t = np.linspace(0,2*np.pi)

x = np.cos(t)
y = np.sin(t)

plt.plot(x,y,c='r')
plt.scatter(X,Y, s=0.5, c='#d5dbdb')
```

### $$\frac{\text{nbr Points}_{\text{in}}}{\text{nbr Points}_{\text{tot}}}  =  \frac{\text{Aire cercle}}{\text{Aire carré}} = \frac{\pi}{4}$$

### $$\Leftrightarrow \pi = 4\frac{\text{nbr Points}_{\text{in}}}{\text{nbr Points}_{\text{tot}}}$$

```python
# Compte le nombre de points dans le cercle
In = np.sqrt(X**2 +Y**2) <= 1

np.count_nonzero(In)
```

```python
# Calcul de la valeur de pi
np.count_nonzero(In)/N*4
```
