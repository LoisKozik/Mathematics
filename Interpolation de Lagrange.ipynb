## Polyn√¥mes de Lagrange

### $$L(X)=\sum_{j=0}^{n} y_j \Bigg ( \prod_{i=0, i \ne j}^{n}  \frac{X-x_i}{x_j-x_i} \Bigg )$$

```python
X = smp.symbols('X')
```

```python
x = np.array([1,9,3])
y = np.array([4,4,5])

n = len(x)
```

```python
def Lagrange(X):
    L = 0
    for j in range(n):
        P = 1
        for i in range(n):
            if i != j:
                P = P * (X - x[i])/(x[j] - x[i])      
        L += y[j] * P

    return simplify(L)    
```

```python
Lagrange(X)
```

```python
a = -20
b = 20
N = 100

T = np.linspace(a,b,N)

plt.plot(T, Lagrange(T))
plt.scatter(x,y, c='orange')
plt.show()
```
